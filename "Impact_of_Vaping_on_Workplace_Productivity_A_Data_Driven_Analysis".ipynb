```python
serie_frecuencia = df_vape['¿Usas vape en la escuela o trabajo? (Si lo usaste alguna vez, marca Si)'].value_counts()

opciones_frecuencia = pd.DataFrame(serie_frecuencia).reset_index()

opciones_frecuencia.columns = ['Opción', 'Incidencia']

opciones_frecuencia = opciones_frecuencia.sort_values(by='Incidencia', ascending=False)

print(opciones_frecuencia)
```

Podemos observar que la gente que da más hits por periodo, o está fumando igual todo el día, es la misma gente que lo hace en la escuela y/o trabajo, por lo que esto lleva a una perdida de tiempo del día util que calcularemos a continuación.

```python
# Definir constantes
puffs = 4000
active_time = 30  # en segundos
refractory_time = 80  # en segundos
hit = 5
labor_time = 28800  # 8 horas de trabajo en segundos


# Cálculo del tiempo total por ciclo de uso (activo + recuperación)
def calcular_total_time(average_lapse, refractory_time):
    return refractory_time + average_lapse

# Cálculo del número de ciclos de "hit" por cada "periodo"
def calcular_hit2(average_lapse, active_time, hit):
    return (average_lapse / active_time) * hit

# Cálculo del tiempo total perdido en fumar durante el día laboral
def calcular_tiempo_perdido(puffs, hit2, total_time, average_vape_lifetime):
    # Calcular la cantidad de lapsos diarios en los cuales se utiliza el vape
    average_daily_lapses = (puffs / hit2) / (average_vape_lifetime / 3)  # Una tercera parte del día es laboral
    # Calcular el tiempo total perdido durante el día
    total_time_loss = (average_lapse * average_daily_lapses) + (refractory_time * average_daily_lapses)
    total_time_loss_m = total_time_loss / 60  # Convertir de segundos a minutos
    return total_time_loss_m, total_time_loss  # Devolver tanto en minutos como en segundos

# Cálculo del tiempo útil laboral después de perder tiempo en fumar
def calcular_tiempo_util(labor_time, total_time_loss):
    return labor_time - total_time_loss

# Cálculo del porcentaje de tiempo perdido en fumar
def calcular_porcentaje_tiempo_perdido(labor_time, total_time_loss):
    percentage_time_loss = (total_time_loss / labor_time) * 100
    return percentage_time_loss

# Ejecutar cálculos
total_time = calcular_total_time(average_lapse, refractory_time)
hit2 = calcular_hit2(average_lapse, active_time, hit)
total_time_loss_m, total_time_loss = calcular_tiempo_perdido(puffs, hit2, total_time, average_vape_lifetime)
percentage_time_loss = calcular_porcentaje_tiempo_perdido(labor_time, total_time_loss)

# Imprimir resultados
print(f'{round(percentage_time_loss)}% del tiempo laboral se pierde en fumar.')
print(f'{round(total_time_loss_m)} minutos del tiempo laboral se pierden en fumar.')
```
